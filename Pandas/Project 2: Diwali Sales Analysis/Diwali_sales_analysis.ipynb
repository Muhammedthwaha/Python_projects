# Exploratory Data Analysis

### **Gender**

```python
ax = sns.countplot(x='Gender', data=df, hue='Gender')  # count plot for gender distribution

for bars in ax.containers:  # add labels to bars
    ax.bar_label(bars)      

plt.show()  # display the plot
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False) 
# total sales based on gender by grouping 'Gender' column and summing up 'Amount' column
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Gender', y='Amount', data=sales_gender, hue='Gender') # bar plot for sales based on gender
plt.show()
```

From above graph we can see that most of the buyers are female and even the purchasing power of the female is greater than male

### **Age**

```python


ax = sns.countplot(x='Age Group', hue= 'Gender', data=df)  
# count plot for age group distribution
# hue parameter is used to add another categorical variable : Here it is 'Gender'
# data parameter is used to specify the dataframe

for bars in ax.containers:   # add labels to bars
    ax.bar_label(bars) 

plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(17,8)})
#colors = ['red', 'green', 'blue', 'orange', 'purple']

sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group')
# bar plot for sales based on age group


plt.show()
```

From above graph we can see that most of the buyers are of age group between 26-35 yrs female

### **State**

```python
# Count plot for state distribution
sns.set(rc={'figure.figsize':(22,5)})

ax = sns.countplot(x='State', hue= 'State', data=df)  

for bars in ax.containers:   # add labels to bars
    ax.bar_label(bars) 

plt.show()
```

```python
# Total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
 # set the figure size for the plot



sns.barplot(x='State', y='Amount', data=sales_state, hue='State')

plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Utter Pradesh, Maharashtra and Karnataka respectively

### **Martial Status**

```python
df.columns
```

```python
ax = sns.countplot(x='Marital_Status', data=df, hue='Marital_Status')  # count plot for marital status distribution
sns.set(rc={'figure.figsize':(10,5)})


for bars in ax.containers:  # add labels to bars
    ax.bar_label(bars)      

plt.show()  # display the plot
```

From above graphs we can see that most of the buyers are married(0) women and they have high purchasing power

### **Occupation**

```python
sns.set(rc={'figure.figsize':(22,5)})

# colors = ['#8da0cb', '#66c2a5', '#fc8d62', '#e78ac3', '#a6d854']


ax = sns.countplot(x='Occupation', data=df, hue='Occupation')  # count plot for occupation distribution

for bars in ax.containers:  # add labels to bars
    ax.bar_label(bars)      

plt.show()  # display the plot
```

```python
# Total amount/sales from Occupation 

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(17,5)})
 # set the figure size for the plot


# colors = ['#8da0cb', '#66c2a5', '#fc8d62', '#e78ac3', '#a6d854']
sns.barplot(x='Occupation', y='Amount', data=sales_state, hue='Occupation')

plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation

### **Product Category**

```python
sns.set(rc={'figure.figsize':(25,8)})

# colors = ['#8da0cb', '#66c2a5', '#fc8d62', '#e78ac3', '#a6d854']


ax = sns.countplot(x='Product_Category', data=df, hue='Product_Category')  # count plot for product category distribution

for bars in ax.containers:  # add labels to bars
    ax.bar_label(bars)      

plt.show()  # display the plot
```

```python
 # Total amount/sales from Product Category

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
 # set the figure size for the plot


# colors = ['#8da0cb', '#66c2a5', '#fc8d62', '#e78ac3', '#a6d854']
sns.barplot(x='Product_Category', y='Amount', data=sales_state, hue='Product_Category')

plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics Category

```python
# Total orders from Product ID

sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
 # set the figure size for the plot

sns.barplot(x='Product_ID', y='Orders', data=sales_state, hue='Product_ID')

plt.show()
```

```python
# Total orders from Product ID (Alternative Method),  (Same thing as above)


sns.set(rc={'figure.figsize':(20,8)})
# set the figure size for the plot

sales_state = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values( ascending=False).plot(kind='bar')

plt.show()
```

### **Conclusion**

Married Women age group 26-35 yrs from UP,Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics Category
